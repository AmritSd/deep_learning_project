```python
import pandas as pd
import numpy as np
import openai
import time
```

```python
yolo = pd.read_csv('clip_int_for_gpt.csv')
yolo = yolo[['prompt', 'clip_inter_yolo']]
yolo
```

```python
yolo.shape
```

```python
openai.api_key = "API KEY HERE"
```

```python
batch_1 = yolo.head(100)
batch_2 = yolo.iloc[100:200]
batch_3 = yolo.iloc[200:300]
batch_4 = yolo.iloc[300:400]
batch_5 = yolo.iloc[400:500]
```

```python
count = 0

def get_words(text):
    response = openai.ChatCompletion.create(
        model = 'gpt-3.5-turbo',
        messages = [
        {'role': 'system', 'content': 'You are an assistant.'},
        {'role': 'user', 'content': f"You are a system that recommends additional words that could describe features of an image described by the user. Respond with a list of five words that are not in the user's message and nothing else. Separate the result by commas (first, second, third, fourth, fifth). The first prompt is: {text}"},
        ]
    )
    time.sleep(3)
    # change count and print to see progress
    global count
    count += 1
    print(count)
    return response['choices'][0]['message']['content']

batch_1['clip_inter_gpt_words'] = batch_1['clip_inter_yolo'].apply(get_words)
```

```python
batch_1.sample(1)
```

```python
# concat all but batch 1
rest = pd.concat([batch_2, batch_3, batch_4, batch_5])

# add column clip_inter_gpt_words
rest['clip_inter_gpt_words'] = ''
```

```python
rest
```

```python
count = 0

# loop through rest and if the row is empty, get words
for i in range(len(rest)):
    if rest.iloc[i]['clip_inter_gpt_words'] == '':
        rest.iloc[i]['clip_inter_gpt_words'] = get_words(rest.iloc[i]['clip_inter_yolo'])
```

```python
# number of rows where clip_inter_gpt_words is ''
rest[rest['clip_inter_gpt_words'] == '']
```

```python
# put all the batches together
gpt = pd.concat([batch_1, rest])
```

```python
clip_inter_gpt = gpt[['prompt', 'clip_inter_gpt_words']]
clip_inter_gpt.to_csv('clip_inter_gpt.csv', index=False)
```

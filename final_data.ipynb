```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
file1 = pd.read_csv('labels_yolo.csv')
file1 = file1[~file1['yolo_prompt'].isna()].reset_index(drop=True)
file1.rename(columns={'first_seven':'clip_inter_two_beam', 'yolo_prompt': 'open_clip_yolo', 'yolo': 'yolo_count'}, inplace=True)
file1 = file1[['prompt', 'open_clip', 'open_clip_yolo', 'yolo_count', 'clip_inter_two_beam']]
```

```python
file2 = pd.read_csv('inter_fine_tune.csv')
file2.rename(columns={'inter_fine_tune': 'clip_inter_fine_tune'}, inplace=True)
```

```python
file3 = pd.read_csv('clip_inter_yolo.csv')
```

```python
file4 = pd.read_csv('open_clip_gpt.csv')
```

```python
file5 = pd.read_csv('clip_inter_gpt.csv')
```

## Open clip: Baseline, Yolo, Yolo + GPT3

```python
open_clip = file1[['prompt', 'open_clip', 'open_clip_yolo', 'yolo_count']]
open_clip['open_clip_yolo'] = np.where(open_clip['yolo_count'] > 1, open_clip['open_clip_yolo'], open_clip['open_clip'])
open_clip = open_clip.merge(file4, on='prompt', how='inner')
open_clip['open_clip_gpt'] = open_clip['open_clip_yolo'] + ' ' + open_clip['open_clip_gpt_words']
open_clip = open_clip[['prompt', 'open_clip', 'open_clip_yolo', 'open_clip_gpt', 'yolo_count']]
open_clip
```

```python
for i in range(18,19):
    print(open_clip.iloc[i, 0])
    print(open_clip.iloc[i, 1])
    print(open_clip.iloc[i, 2])
    print(open_clip.iloc[i, 3])
    print()
```

## Clip Interrogator: Baseline, Fine Tuned, Fine Tuned + Yolo, Fine Tuned + Yolo + GPT3

```python
clip_inter = file1[['prompt', 'clip_inter_two_beam']]
clip_inter = clip_inter.merge(file2, on='prompt', how='inner')
clip_inter = clip_inter.merge(open_clip[['prompt', 'yolo_count']], on='prompt', how='inner')
clip_inter = clip_inter.merge(file3, on='prompt', how='left')
clip_inter['clip_inter_yolo'] = np.where(clip_inter['clip_inter_yolo'].isna(), clip_inter['clip_inter_fine_tune'], clip_inter['clip_inter_yolo'])
clip_inter = clip_inter.merge(file5, on='prompt', how='left')
clip_inter['clip_inter_gpt'] = clip_inter['clip_inter_yolo'] + ' ' + clip_inter['clip_inter_gpt_words']
clip_inter = clip_inter[['prompt', 'clip_inter_two_beam', 'clip_inter_fine_tune', 'clip_inter_yolo', 'clip_inter_gpt']]
clip_inter
```

```python
for i in range(496,497):
    print(clip_inter.iloc[i, 0])
    print(clip_inter.iloc[i, 1])
    print(clip_inter.iloc[i, 2])
    print(clip_inter.iloc[i, 3])
    print(clip_inter.iloc[i, 4])
    print()
```

```python
data = open_clip.merge(clip_inter, on='prompt', how='inner')

data.to_csv('final.csv', index=False)
```
